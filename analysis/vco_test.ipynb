```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import pandas as pd 
import seaborn as sns
 
df = pd.read_csv('results_vco.csv')
```

```python
sns.scatterplot(x='vcont', y='freq', hue='corner', data=df.query('s0==0 and s1==0'))
plt.show()
```

```python
fig, ax = plt.subplots()
sns.scatterplot(x='vcont', y='freq', hue='corner', data=df.query('s0==0 and s1==0'), s=8.0)
ax.xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: '{:,.2f}'.format(x) + 'V'))
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: '{:,.2f}'.format(x/1000000) + 'MHz'))
ax.set_box_aspect(2.0)
fig.set_dpi(200)
plt.show()
```
