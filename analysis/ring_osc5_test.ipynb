```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import pandas as pd 
import seaborn as sns
 
df3 = pd.read_csv('results_ring_osc3.csv')
df5 = pd.read_csv('results_ring_osc5.csv')
df7 = pd.read_csv('results_ring_osc7.csv')
df9 = pd.read_csv('results_ring_osc9.csv')
df = pd.concat([df3, df5, df7, df9])
df = df.assign(period=lambda df: 1.0/df.freq)
```

```python
fig, ax = plt.subplots()
sns.scatterplot(x='vcont', y='freq', hue='corner', data=df, s=8.0)
ax.xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: '{:,.2f}'.format(x) + 'V'))
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: '{:,.2f}'.format(x/1000000) + 'MHz'))
ax.set_box_aspect(2.0)
fig.set_dpi(200)
plt.show()
```

```python
fig, ax = plt.subplots()
sns.scatterplot(ax=ax, x='vcont', y='period', hue='corner', data=df, s=8.0)
#plt.ylim(0, 1e-8)
ax.xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: '{:,.2f}'.format(x) + 'V'))
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: '{:,.2f}'.format(x * 1000000000) + 'ns'))
ax.set_box_aspect(2.0)
fig.set_dpi(200)
plt.show()
```
