```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import pandas as pd 
import seaborn as sns
 
df = pd.read_csv('results_ring_osc3.csv')
df=df.assign(period=lambda df: 1.0/df.freq)
```

```python
fig, ax = plt.subplots()
sns.scatterplot(x='vcont', y='freq', hue='corner', data=df)
ax.xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: '{:,.2f}'.format(x) + 'V'))
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: '{:,.2f}'.format(x/1000000) + 'MHz'))
plt.show()
```

```python
fig, ax = plt.subplots()
sns.scatterplot(ax=ax, x='vcont', y='period', hue='corner', data=df)
#plt.ylim(0, 1e-8)
ax.xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: '{:,.2f}'.format(x) + 'V'))
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: '{:,.2f}'.format(x * 1000000000) + 'ns'))
plt.show()
```
